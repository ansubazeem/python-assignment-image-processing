```python
#For convrting image into array 
import matplotlib.pyplot as plt
import matplotlib.image as img
import os
import glob as gb
data_path=os.path.join (r"D:\pictures","*.png")
files=gb.glob(data_path)
for f2 in files:
    z=img.imread(f2)
    print(z)
import pickle
pickle.dump( z, open( "matplotlibarrays.p", "wb" ) )

```
